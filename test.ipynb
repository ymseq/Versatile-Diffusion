```python
import json
import os
from PIL import Image

import jittor as jt
from jittor.dataset.dataset import Dataset
import torchvision.transforms as transforms
```

```python
image_path = r'./data/train/images/2125216241_5b265a2fbc.jpg'
size = 256
```

```python
img = Image.open(image_path)
img.load()
img = img.convert("RGB")
```

```python
transform = transforms.Compose([
    transforms.Resize(size),
    transforms.CenterCrop(size),
    transforms.ToTensor(),  # -> jt.Var, CHW, float32, [0,1]
    transforms.Normalize(mean=[0.5, 0.5, 0.5],
                            std=[0.5, 0.5, 0.5]),
])
transformed_img = transform(img)
```

```python
import json

with open("./data/train/meta.jsonl", "r", encoding="utf-8") as f:
    for i, line in enumerate(f, 1):
        line = line.strip()
        if not line:
            continue
        try:
            json.loads(line)
        except json.JSONDecodeError as e:
            print("BAD LINE:", i)
            print("ERROR:", e)
            print("CONTENT REPR:", repr(line[:300]))
            raise
```
